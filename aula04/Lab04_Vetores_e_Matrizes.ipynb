## 1. Vetores (Arrays 1D)

Um **vetor** (ou array unidimensional) é uma sequência contígua de elementos na memória, todos do mesmo tipo.
No Python, a lista (`list`) simula muito bem um vetor de alocação dinâmica.

### Alocação de Memória
Cada elemento ocupa um bloco fixo de memória. O acesso por índice é sempre $O(1)$ — custo constante, independente do tamanho.

```python
# Criando um vetor de inteiros com N elementos
N = 10
vetor: list[int] = list(range(N))

print(f"Vetor criado:  {vetor}")
print(f"Tamanho:       {len(vetor)} elementos")
print(f"Memória usada: {sys.getsizeof(vetor)} bytes")
```

### 1.1 Acesso por Índice — $O(1)$

O acesso direto através do índice é a operação mais rápida possível: o custo **não cresce** com o tamanho do vetor.

```python
def acessar_elemento(vetor: list, indice: int) -> int:
    """
    Retorna o elemento na posição 'indice' do vetor.
    Complexidade: O(1) — acesso direto por endereço de memória.
    """
    return vetor[indice]


# Teste
print(f"Elemento no índice 5: {acessar_elemento(vetor, 5)}")
```

### 1.2 Inserção no Fim — $O(1)$ amortizado

Adicionar no **final** é barato: nenhum elemento precisa ser deslocado.

```python
def inserir_no_fim(vetor: list, valor: int) -> list:
    """
    Insere 'valor' no final do vetor.
    Complexidade: O(1) amortizado.
    """
    vetor.append(valor)
    return vetor


vetor_teste = list(range(5))          # [0, 1, 2, 3, 4]
inserir_no_fim(vetor_teste, 99)
print(f"Após inserção no fim: {vetor_teste}")
```

### 1.3 Inserção no Início / Meio — $O(n)$

Inserir em qualquer posição que **não seja o final** exige deslocar todos os elementos à direita. O custo cresce linearmente com o tamanho do vetor.

```python
def inserir_em_posicao(vetor: list, valor: int, posicao: int) -> list:
    """
    Insere 'valor' na 'posicao' indicada do vetor.
    Complexidade: O(n) — desloca todos os elementos à direita da posição.
    """
    vetor.insert(posicao, valor)
    return vetor


vetor_teste = list(range(5))          # [0, 1, 2, 3, 4]
inserir_em_posicao(vetor_teste, 42, 0)   # insere no início
print(f"Após inserção no início: {vetor_teste}")
```

### 1.4 Remoção — $O(1)$ vs $O(n)$

Assim como na inserção:
* Remover do **final** → $O(1)$
* Remover de qualquer outra posição → $O(n)$ (desloca elementos à esquerda)

```python
def remover_do_fim(vetor: list) -> int:
    """
    Remove e retorna o último elemento do vetor.
    Complexidade: O(1).
    """
    return vetor.pop()


def remover_de_posicao(vetor: list, posicao: int) -> int:
    """
    Remove e retorna o elemento na 'posicao' indicada.
    Complexidade: O(n) — desloca os elementos restantes.
    """
    return vetor.pop(posicao)


dados = [10, 20, 30, 40, 50]
removido = remover_do_fim(dados)
print(f"Removido do fim: {removido} → vetor restante: {dados}")

dados = [10, 20, 30, 40, 50]
removido = remover_de_posicao(dados, 0)  # remove do início
print(f"Removido do início: {removido} → vetor restante: {dados}")
```

## 🧪 Exemplo Aplicado 1: Fila de Atendimento (Vetor)

Vamos simular uma **fila de atendimento** de um banco.
Cada operação mostra o estado do vetor e o custo teórico, para você visualizar o efeito do deslocamento de memória.

```python
# Estado inicial: três clientes já aguardam
fila = ["Carlos", "Beatriz", "Arnaldo"]
print(f"Fila inicial:                    {fila}")
print(f"Memória:                         {sys.getsizeof(fila)} bytes\n")

# 1) Novo cliente entra no FIM da fila  →  O(1)
inserir_no_fim(fila, "Daniela")
print(f"[O(1)] Daniela entrou no fim:    {fila}")

# 2) Cliente VIP é inserido na FRENTE  →  O(n): todos recuam uma posição
inserir_em_posicao(fila, "👑 Elisa (VIP)", 0)
print(f"[O(n)] Elisa entrou na frente:   {fila}")

# 3) Primeiro cliente é atendido (removido do início)  →  O(n): todos avançam
atendido = remover_de_posicao(fila, 0)
print(f"[O(n)] '{atendido}' foi atendido.")
print(f"       Fila restante:             {fila}")

# 4) Último cliente desiste e vai embora  →  O(1): sem deslocamento
desistiu = remover_do_fim(fila)
print(f"[O(1)] '{desistiu}' desistiu.")
print(f"       Fila restante:             {fila}")

# 5) Acesso direto: consultando quem está na posição 0  →  O(1)
print(f"\n[O(1)] Próximo da fila (índice 0): {acessar_elemento(fila, 0)}")
```

> **💡 Perceba:** inserir/remover no **início** exige deslocar *todos* os outros elementos.
> Para uma fila com 10.000 clientes, isso significa 10.000 movimentações de memória a cada atendimento.
> É por isso que estruturas como **Deque** ou **Listas Encadeadas** (próximas aulas) surgem como alternativas.

## 2. Matrizes (Arrays 2D)

Uma **matriz** é um vetor de vetores — cada linha é ela mesma um vetor. Em Python, representamos com listas de listas.

$$
M = \begin{bmatrix} m_{00} & m_{01} & m_{02} \\ m_{10} & m_{11} & m_{12} \\ m_{20} & m_{21} & m_{22} \end{bmatrix}
$$

```python
def criar_matriz(linhas: int, colunas: int, valor_padrao: int = 0) -> list[list[int]]:
    """
    Cria e retorna uma matriz de dimensão (linhas x colunas) preenchida com 'valor_padrao'.
    Complexidade: O(linhas * colunas).
    """
    return [[valor_padrao] * colunas for _ in range(linhas)]


def imprimir_matriz(matriz: list[list[int]]) -> None:
    """Imprime a matriz linha por linha de forma legível."""
    for linha in matriz:
        print(linha)


M = criar_matriz(3, 3)
print("Matriz 3x3 zerada:")
imprimir_matriz(M)
```

### 2.1 Acesso e Modificação — $O(1)$

Ler ou escrever em qualquer célula `M[linha][coluna]` é sempre $O(1)$.

```python
def preencher_identidade(n: int) -> list[list[int]]:
    """
    Cria e retorna uma Matriz Identidade n x n.
    Complexidade: O(n²) — percorre cada célula uma vez.
    """
    matriz = criar_matriz(n, n)
    for i in range(n):
        matriz[i][i] = 1   # acesso O(1) por posição
    return matriz


identidade = preencher_identidade(4)
print("Matriz Identidade 4x4:")
imprimir_matriz(identidade)
```

### 2.2 Percurso Completo da Matriz — $O(n^2)$

Para visitar **todos** os elementos de uma matriz $n \times n$ precisamos de dois loops aninhados.
O custo é **quadrático**: duplicar $n$ quadruplica o trabalho.

```python
def somar_todos_elementos(matriz: list[list[int]]) -> int:
    """
    Retorna a soma de todos os elementos da matriz.
    Complexidade: O(linhas * colunas) = O(n²) para matrizes quadradas.
    """
    total = 0
    for linha in matriz:
        for elemento in linha:
            total += elemento
    return total


# Teste com a identidade 4x4 (soma deve ser 4)
soma = somar_todos_elementos(identidade)
print(f"Soma de todos os elementos da identidade 4x4: {soma}")
```

### 2.3 Transposição de Matriz — $O(n^2)$

A transposta de $M$ troca linhas por colunas: $M^T[i][j] = M[j][i]$.

```python
def transpor_matriz(matriz: list[list[int]]) -> list[list[int]]:
    """
    Retorna a transposta da matriz de entrada.
    Complexidade: O(linhas * colunas).
    """
    linhas = len(matriz)
    colunas = len(matriz[0])
    transposta = criar_matriz(colunas, linhas)
    for i in range(linhas):
        for j in range(colunas):
            transposta[j][i] = matriz[i][j]
    return transposta


M_original = [
    [1, 2, 3],
    [4, 5, 6]
]

M_T = transpor_matriz(M_original)

print("Matriz original (2x3):")
imprimir_matriz(M_original)

print("\nTransposta (3x2):")
imprimir_matriz(M_T)
```

## 🧪 Exemplo Aplicado 2: Tabela de Notas (Matriz)

Vamos usar uma **tabela de notas** de alunos em disciplinas como exemplo de matriz.

| | Estruturas | Algoritmos | Banco de Dados |
|---|---|---|---|
| **Ana**    | 9.5 | 8.0 | 7.5 |
| **Bruno**  | 6.0 | 7.5 | 8.5 |
| **Carla**  | 8.0 | 9.0 | 9.5 |
| **Daniel** | 7.0 | 6.5 | 7.0 |

```python
# Representação da tabela como matriz (lista de listas)
alunos = ["Ana", "Bruno", "Carla", "Daniel"]
disciplinas = ["Estruturas", "Algoritmos", "Banco de Dados"]

notas: list[list[float]] = [
    [9.5, 8.0, 7.5],   # Ana    — linha 0
    [6.0, 7.5, 8.5],   # Bruno  — linha 1
    [8.0, 9.0, 9.5],   # Carla  — linha 2
    [7.0, 6.5, 7.0],   # Daniel — linha 3
]

# 1) Acesso direto O(1): nota de Carla em Algoritmos  →  linha 2, coluna 1
print("=== Acesso por índice O(1) ===")
print(f"Nota de {alunos[2]} em {disciplinas[1]}: {notas[2][1]}\n")

# 2) Percurso O(n²): média geral de cada aluno
print("=== Média por aluno — O(n²) ===")
for i, aluno in enumerate(alunos):
    media = sum(notas[i]) / len(notas[i])
    print(f"  {aluno:<8}: {media:.2f}")

# 3) Percurso O(n²): média da turma por disciplina
print("\n=== Média por disciplina — O(n²) ===")
for j, disciplina in enumerate(disciplinas):
    media = sum(notas[i][j] for i in range(len(alunos))) / len(alunos)
    print(f"  {disciplina:<16}: {media:.2f}")

# 4) Transposição: visão por disciplina × aluno
print("\n=== Transposta: disciplinas × alunos ===")
notas_T = transpor_matriz(notas)
print(f"  {'':18}", end="")
for aluno in alunos:
    print(f"{aluno:>8}", end="")
print()
for j, disciplina in enumerate(disciplinas):
    print(f"  {disciplina:<18}", end="")
    for nota in notas_T[j]:
        print(f"{nota:>8.1f}", end="")
    print()
```

> **💡 Perceba:** com apenas `notas[linha][coluna]` acessamos qualquer célula em $O(1)$.
> Mas para calcular médias ou transpor, precisamos **visitar cada célula**, resultando em $O(n \times m)$.
> Para uma tabela com 1.000 alunos × 50 disciplinas isso são **50.000 operações** só para calcular as médias.

## 3. Benchmarking: Crescimento da Complexidade

Vamos medir empiricamente como o tempo de execução cresce com o tamanho $n$:

| Operação | Complexidade esperada |
|---|---|
| Acesso por índice | $O(1)$ |
| Inserção no fim (vetor) | $O(1)$ amortizado |
| Inserção no início (vetor) | $O(n)$ |
| Percurso da matriz | $O(n^2)$ |

```python
tamanhos = [100, 500, 1000, 5000, 10000]

# --- Inserção no INÍCIO (O(n)) vs no FIM (O(1)) ---
print("=" * 55)
print(f"{'N':>7} | {'Inserção no Início (s)':>23} | {'Inserção no Fim (s)':>20}")
print("-" * 55)

for n in tamanhos:
    dados = list(range(n))

    # Inserção no início
    inicio = perf_counter()
    dados.insert(0, -1)
    fim = perf_counter()
    tempo_inicio = fim - inicio

    dados = list(range(n))

    # Inserção no fim
    inicio = perf_counter()
    dados.append(-1)
    fim = perf_counter()
    tempo_fim = fim - inicio

    print(f"{n:>7} | {tempo_inicio:>23.9f} | {tempo_fim:>20.9f}")

print()

# --- Percurso de Matriz n x n (O(n²)) ---
print("=" * 40)
print(f"{'N':>6} | {'Soma Matriz NxN (s)':>20}")
print("-" * 40)

for n in [50, 100, 200, 400, 800]:
    matriz = [[i * n + j for j in range(n)] for i in range(n)]

    inicio = perf_counter()
    somar_todos_elementos(matriz)
    fim = perf_counter()

    print(f"{n:>6} | {fim - inicio:>20.9f}")
```

## 4. Desafio: Multiplicação de Matrizes — $O(n^3)$

A multiplicação clássica entre duas matrizes $n \times n$ possui três loops aninhados, gerando complexidade **cúbica**.
> ⚠️ **Atenção:** Para $n = 500$, são executadas $500^3 = 125.000.000$ operações. Rode com cuidado!

```python
def multiplicar_matrizes(
    A: list[list[float]], B: list[list[float]]
) -> list[list[float]]:
    """
    Multiplica as matrizes A e B e retorna a matriz resultante C = A x B.
    Requer que as dimensões sejam compatíveis: colunas(A) == linhas(B).
    Complexidade: O(n³) para matrizes n x n.
    """
    n_linhas_A = len(A)
    n_colunas_A = len(A[0])
    n_colunas_B = len(B[0])

    C = criar_matriz(n_linhas_A, n_colunas_B)

    for i in range(n_linhas_A):
        for j in range(n_colunas_B):
            for k in range(n_colunas_A):
                C[i][j] += A[i][k] * B[k][j]
    return C


# Benchmark de multiplicação de matrizes quadradas
print("=" * 38)
print(f"{'N':>5} | {'Multiplicação NxN (s)':>20}")
print("-" * 38)

for n in [10, 30, 50, 80, 100]:
    A = [[1.0] * n for _ in range(n)]
    B = [[1.0] * n for _ in range(n)]

    inicio = perf_counter()
    multiplicar_matrizes(A, B)
    fim = perf_counter()

    print(f"{n:>5} | {fim - inicio:>20.6f}")
```
