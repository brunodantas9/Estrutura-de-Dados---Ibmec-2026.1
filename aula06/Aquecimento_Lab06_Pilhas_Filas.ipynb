### Situação 2: Como o Python faz isso?
A classe nativa `list` do Python já possui comandos embutidos para adicionar e remover dados. Vamos testá-la.
1. Crie uma lista vazia chamada `historico`.
2. Use o comando `.append()` para adicionar as strings: `"Passo 1"`, `"Passo 2"` e `"Passo 3"`.
3. Use o comando `.pop()` (sem nenhum argumento) e guarde o resultado numa variável chamada `removido`. Imprima a lista e o item removido.

```python
# Escreva seu código aqui
historico = []
historico.append("Passo 1")
historico.append("Passo 2")
historico.append("Passo 3")

removido = historico.pop()

print("Item desfeito:", removido)
print("Restante na memória:", historico)
```

```python
#como você faria isso usando um laço ?
historico = []

for i in range (1, 4):
    historico.append(f"Passo {i}")

removido = historico.pop()

print("Item desfeito:", removido)
print("Restante na memória:", historico)

```
